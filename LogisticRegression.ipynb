```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
# Read the bank csv file and provide summary
bankdf = pd.read_csv('/content/Bank.csv')
print(bankdf.info())
bankdf.head(5)
```

```python
# Defining dependent and independent variables
X2 = bankdf[['Age', 'Experience', 'Income', 'Education']] # independent (X) variable.
Y2 = bankdf['Personal Loan'] #class (Y) variable
```

```python
# c) Creating the training and testing datasets
X_train2, X_test2, Y_train2, Y_test2 = train_test_split(X2, Y2, test_size=0.2, random_state=42)
```

```python
# d) Using the Logistic Regression model
bankmodel = LogisticRegression()
bankmodel.fit(X_train2, Y_train2)
```

```python
# Model Evaluation
Y_pred2 = bankmodel.predict(X_test2)
```

```python
# Result
accuracy = accuracy_score(Y_test2, Y_pred2)
f1 = f1_score(Y_test2, Y_pred2)
conf_matrix = confusion_matrix(Y_test2, Y_pred2)

print("Accuracy:", np.round(accuracy*100,2),"%")
print("F1 Score:", f1)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Creating a new data points for X and making the predictions
new_data = pd.DataFrame({
    'Age': [25, 37, 50, 63, 28],
    'Experience': [1, 10, 20, 15, 5],
    'Income': [50, 180, 180, 70, 120],
    'Education': [0, 3, 1, 2, 3]
})
```

```python
predictions = bankmodel.predict(new_data)
print("Predictions for the new data points:")
print(predictions)
```
