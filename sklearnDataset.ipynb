```python
# Import pandas (for data manipulation and analysis)
import pandas as pd

#import datasets which holds all the datasets.
from sklearn import datasets
```

```python
#dir(datasets) list directories in dataset

# Code to list all the directories that start with load
# List of codes to display available dataset in sklearn
for data in dir(datasets):
    if data.startswith("load"):
        print(data)
```

```python
# loading the diabetes dataset and displaying the target and feautures
# Load the diabetes dataset
diabetes = datasets.load_diabetes()

diabetes.keys()
```
